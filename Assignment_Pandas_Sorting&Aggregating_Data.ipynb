#### `Consider the following dataframe`

---

```python
sample_df = pd.DataFrame({
    'id'   :  [1, 2, 3, 4, 5],
    'name' :  [ 'Akash', 'Akansha', 'Naman', 'Pulkit', 'Anand'],
    'marks':  [ 22, 23, 43, 41, 12]
})
sample_df
```

#### Sort the  dataframe by `marks` column in descending order

---

```python
# WRITE YOUR CODE HERE
sample_df.sort_values(by='marks', ascending=False)
```

#### Consider another dataframe city
---

```python
city = pd.DataFrame({
    'name' : ['Akash', 'Akansha', 'Pulkit', 'Anand', 'Naman'],
    'city' : ['Delhi', 'Gurugram', 'Gurugram', 'Delhi', 'Noida']
})
city
```

***Use the dataframe `sample_df` and `city` and find out the marks of students who live in Gurugram.***

---

```python
### Write your code here
merged_df = pd.merge(sample_df, city, on='name', how='left')
```

```python
merged_df.head()
```

```python
marks_students_Gurugram = merged_df[merged_df['city'] == 'Gurugram']['marks']
marks_students_Gurugram
```

####Answer to the question:
The marks of the students living Gurugram are: 23 and 41

#### `READ THE BIG MART SALES DATA`

---

```python
big_mart_data = pd.read_csv('/content/drive/My Drive/Assignment3/dataset/big_mart_sales.csv')
```

```python
big_mart_data.head()
```

#### Find out the number of unique `Item_Type` in each of the `Outlet_Identifier`

---

```python
# WRITE YOUR CODE HERE
big_mart_data.groupby('Outlet_Identifier')['Item_Type'].nunique()
```

#### Find out the average `Item_Outlet_Sales` for each category of `Outlet_Size`

---

```python
# WRITE YOUR CODE HERE
big_mart_data.groupby('Outlet_Size')['Item_Outlet_Sales'].mean()
```

#### Select only the categorical variables from the dataframe and use `get_dummies` function to convert those into the numerical ones.

---

```python
# WRITE YOUR CODE HERE
categorical_df = big_mart_data.select_dtypes(include=['object'])
categorical_df.head()
```

```python
categorical_df.shape
```

```python
categorical_df = pd.get_dummies(categorical_df)
categorical_df.head()
```

```python
categorical_df.shape
```

```python
# As you can see the usage of the get_dummies function on the categorical_df has increased the number of columns from 7 to 1,600!
```
